```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(10, 10), dpi=360)
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
gossip_full_success = data.loc[(data['topology'] == 'full') & (data['Deathprob'] == 0.0) \
                              & (data['algo'] == 'gossip')].reset_index(drop=True).sort_values(by=['n'])
gossip_full_success
```

```python
gossip_line_success = data.loc[(data['topology'] == 'line') & (data['Deathprob'] == 0.0) \
                              & (data['algo'] == 'gossip')].reset_index(drop=True).sort_values(by=['n'])
gossip_line_success
```

```python
gossip_grid2d_success = data.loc[(data['topology'] == 'grid_2d') & (data['Deathprob'] == 0.0) \
                              & (data['algo'] == 'gossip')].reset_index(drop=True).sort_values(by=['n'])
gossip_grid2d_success
```

```python
gossip_grid3d_success = data.loc[(data['topology'] == 'grid_3d') & (data['Deathprob'] == 0.0) \
                              & (data['algo'] == 'gossip')].reset_index(drop=True).sort_values(by=['n'])
gossip_grid3d_success
```

```python
# plot lines
plt.plot(gossip_full_success['n'], gossip_full_success['time(microseconds)']/1000, label = gossip_full_success['topology'][0], marker='o')
plt.plot(gossip_line_success['n'], gossip_line_success['time(microseconds)']/1000, label = gossip_line_success['topology'][0], marker='v')
plt.plot(gossip_grid2d_success['n'], gossip_grid2d_success['time(microseconds)']/1000, label = gossip_grid2d_success['topology'][0], marker='^')
plt.plot(gossip_grid3d_success['n'], gossip_grid3d_success['time(microseconds)']/1000, label = gossip_grid3d_success['topology'][0], marker='*')
plt.legend()
plt.xlabel("Number of Actors")
plt.ylabel("Time in milliseconds")
plt.savefig('gossip_runtime')
plt.rcParams["figure.figsize"] = (10,8)
plt.show()
```
