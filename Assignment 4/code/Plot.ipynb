```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
figure(figsize=(10, 10), dpi=360)
```

```python
figure(figsize=(10, 10), dpi=360)
data = [{"4",21},
 {"7",18},
 {"9",19},
 {"1",19},
 {"3",17},
 {"6",14},
 {"10",19},
 {"8",15},
 {"2",17},
 {"5",14}]
```

```python
x = []
y = []

for datum in data:
    datum = list(datum)
    if type(datum[0]) == str:
        x.append(int(datum[0]))
        y.append(datum[1])
    else:
        x.append(int(datum[1]))
        y.append(datum[0])

df = pd.DataFrame([x, y], index=['users', 'tweets']).T
df = df.sort_values(by=['users']).reset_index(drop=True)
```

```python
# plot lines
plt.plot(df['users'], df['tweets'], label = 'Tweets per user')#, marker='o')
plt.legend()
plt.xlabel("User")
plt.ylabel("Number of Tweets")
#plt.savefig('gossip_runtime')
plt.rcParams["figure.figsize"] = (10,8)
plt.show()
```
